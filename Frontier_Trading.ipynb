```python
import requests

def get_sma(stock_ticker, timespan='day', adjusted=True, window=50, series_type='close', order='desc', limit=10, api_key= polygon_key ):
    url = f'https://api.polygon.io/v1/indicators/sma/{stock_ticker}'
    
    params = {
        'timespan': timespan,
        'adjusted': str(adjusted).lower(),
        'window': window,
        'series_type': series_type,
        'order': order,
        'limit': limit,
        'apiKey': api_key
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        return {'error': f'Failed to fetch data: {response.status_code}'}

# Example usage
stock_ticker = 'AAPL'  # Replace with the stock ticker you're interested in
api_key = polygon_key  # Replace with your Polygon.io API key

sma_data = get_sma(stock_ticker, api_key=api_key)
print(sma_data)
```

```python
import pandas as pd
import datetime

# Initialize the Polygon client

# Define ticker list
ticker_list = ["AAPL", "MSFT", "TSLA", "GOOG"]  # Replace with your ticker list

# Define the time range for the last 3 years
end_date = datetime.datetime.today().strftime('%Y-%m-%d')
start_date = (datetime.datetime.today() - datetime.timedelta(days=3*365)).strftime('%Y-%m-%d')

# Function to fetch daily stock data
def fetch_stock_data(symbol, start_date, end_date):
    aggs = client.get_aggs(symbol, 1, 'day', start_date, end_date)
    df = pd.DataFrame(aggs, columns=["timestamp", "open", "high", "low", "close", "volume", "vwap"])
    df["timestamp"] = pd.to_datetime(df["timestamp"], unit="ms")
    df.set_index("timestamp", inplace=True)
    return df

# Function to fetch financials
def fetch_financials(symbol):
    financials = client.get_ticker_details(symbol)
    return financials

# Function to fetch ticker news
def fetch_ticker_news(symbol, limit=5):
    news = []
    for i, n in enumerate(client.list_ticker_news(symbol, limit=limit)):
        news.append(n)
        if i == limit - 1:
            break
    return news

# Data retrieval for each ticker
stock_dataframes = {}
financial_data = {}
news_data = {}

for ticker in ticker_list:
    # Fetch stock data
    stock_dataframes[ticker] = fetch_stock_data(ticker, start_date, end_date)
    
    # Fetch financial data
    financial_data[ticker] = fetch_financials(ticker)
    
    # Fetch ticker news
    news_data[ticker] = fetch_ticker_news(ticker)

# Example to save dataframes as CSVs (optional)
for ticker, df in stock_dataframes.items():
    df.to_csv(f"{ticker}_stock_data.csv")

# Example to print the first few rows of the stock data for each ticker (optional)
for ticker, df in stock_dataframes.items():
    print(f"Stock data for {ticker}:")
    print(df.head())

# Example to print the financials and news for each ticker (optional)
for ticker in ticker_list:
    print(f"Financials for {ticker}:")
    print(financial_data[ticker])
    print(f"News for {ticker}:")
    print(news_data[ticker])

# You can also save the financial and news data to CSV or pickle files if needed
# For example, to save the news data to a CSV for each ticker:
for ticker, news in news_data.items():
    pd.DataFrame(news).to_csv(f"{ticker}_news.csv", index=False)
```
